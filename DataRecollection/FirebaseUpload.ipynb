```python
import firebase_admin
from firebase_admin import credentials, storage
import pandas as pd
import os

if not firebase_admin._apps:
    # Initialize Firebase Admin SDK
    cred = credentials.Certificate('FIreBaseKEY.json')
    firebase_admin.initialize_app(cred, {
        'storageBucket': 'fgc-recommendator.appspot.com'
    })

# Function to upload a file and get its URL
def upload_image(image_path):
    bucket = storage.bucket()
    blob = bucket.blob(os.path.basename(image_path))
    blob.upload_from_filename(image_path)
    blob.make_public()
    return blob.public_url

```

```python
import os
import pandas as pd
dataset_path = 'dataset'
df  = pd.DataFrame(columns=['Games', 'Image Path'])

# Iterate through all directories in the dataset
for root, dirs, files in os.walk(dataset_path):
    for file in files:
        # Get the directory name
        directory_name = os.path.basename(root)
        
        # Get the file extension
        file_name, file_extension = os.path.splitext(file)
        
        # Create the new file name
        new_file_name = directory_name + file_extension
        
        # Get the current file path
        current_file_path = os.path.join(root, file)
        
        # Get the new file path
        new_file_path = os.path.join(root, new_file_name)
        
        # Rename the file
        os.rename(current_file_path, new_file_path)
        
        # Append the new row to the dataframe
        df = pd.concat([df, pd.DataFrame({'Games': [directory_name], 'Image Path': [new_file_path]})], ignore_index=True)

df.to_csv('GamesPaths.csv', index=False)

```

```python
Games_Paths = pd.read_csv('GamesPaths.csv')
Games= pd.read_csv('../Games.csv')
Games["Games"] = Games["Games"].str.replace(':', '')

for index, row in Games.iterrows():
    game = row['Games']
    game_path = Games_Paths.loc[Games_Paths['Games'] == game, 'Image Path']
    try :
        game_path = game_path[game_path.index[0]]
    except:
        print("Game"+game+" not found")
        continue
    print(game_path)
    # game_path = game_path[game_path.index[0]]
    game_url = upload_image(game_path)
    Games.loc[index, 'ImageLink'] = game_url



Games.to_csv('Games.csv', index=False)

```
